### Step 1: Initialize Flask App

```python
from flask import Flask
app = Flask(__name__)
```

### Step 2: Configure your Database

```python
from flask_sqlalchemy import SQLAlchemy
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///themepark.db'
db = SQLAlchemy(app)
```

### Step 3: Define Models


This is where we set up:the structure of the database, the tables (also called entities), how they are related to each other, the data dictionary, the field names (also called attributes), field length, field data type and any other useful information. 
The database will contain a table of customers, staff, events and bookings.

```python
class User(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    email = db.Column(db.String(255), unique=True)
    role = db.Column(db.String(50), default="student")
```

## 4. Going Live: Deployment Steps

### Step 1: Containerization


```python
%%bash
cat > Dockerfile <<'EOF'
FROM python:3.11-slim
WORKDIR /app
COPY requirements.txt .
RUN pip install -r requirements.txt
COPY . .
CMD ["gunicorn", "-b", "0.0.0.0:8080", "app:app"]
EOF
```

### Step 2: Fly.io Config


```python
%%bash
cat > fly.toml <<'EOF'
app = "themepark-pwa"
primary_region = "syd"
[build]
  dockerfile = "Dockerfile"
EOF
```

### Step 3: CI/CD
- `.github/workflows/ci.yml` â†’ linting, tests, coverage.
- `.github/workflows/deploy.yml` â†’ auto-deploy to Heroku or Fly.io.

### Step 4: Production Modifications
- Switch DB URI to Postgres (`SQLALCHEMY_DATABASE_URI`).
- Set `SESSION_COOKIE_SECURE=True`.
- Use environment variables for secrets (`SECRET_KEY`, OAuth keys).
- Configure HTTPS (TLS via Fly.io or Certbot).
- Add logging to external service if scaling.
